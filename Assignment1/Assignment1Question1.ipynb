```python
from nltk.corpus import gutenberg 
gutenberg.fileids()
austen=gutenberg.words('austen-persuasion.txt')
print(austen)
print(len(austen))
print(len(set(austen)))
wordinausten={}
for word1 in austen:
    if word1 in wordinausten:
        wordinausten[word1] +=1
    else:
        wordinausten[word1]=1
print (wordinausten)
```

```python
import operator
sorted_x = sorted(wordinausten.items(), key=operator.itemgetter(1))
print(sorted_x)
print(sorted_x[-10:])
```
