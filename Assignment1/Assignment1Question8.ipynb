```python
import nltk 
from nltk.corpus import stopwords
from nltk.corpus import nps_chat  
file = nps_chat.fileids()
file
```

```python
chat_1 = nps_chat.words()
chat_1[0:5]
chat_1 = [x.lower() for x in chat_1]
filtered_chat = [word for word in chat_1 if (word.lower()) not in stopwords.words('english')]
filtered_chat
```

```python
import re
filter_punctuation = re.compile('.*[A-Za-z].*') 
filtered_webtext_punctuation = [w for w in filtered_chat if filter_punctuation.match(w)]
webchat_wordlist=set(filtered_webtext_punctuation)
webchat_wordlist
```

```python
 #importing webtext from nltk corpus
from nltk.corpus import words
```

```python
word_file = words.fileids()
word_file
```

```python
word_list = []  
for w in word_file:           
        word1= words.words(w)
        word_list.extend(word1)
word_list = [x.lower() for x in word_list]
word_list
```

```python
unusual=[x for x in webchat_wordlist if x not in word_list]
```

```python
# Unusual words people speak in chat 
print("Freq of unusual words: ",len(unusual))
print("Unusual words used in web chat are:")
for unusual_1 in sorted(unusual): 
    print(unusual_1)
```
