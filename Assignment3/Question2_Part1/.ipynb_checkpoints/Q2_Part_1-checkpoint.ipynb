```python
import pandas as pd
```

```python
df=pd.read_csv(r'employee_compensation.csv')
```

```python
df.head()
```

```python
df_group=df.groupby('Organization Group').agg({'Total Compensation':sum})
```

```python
g = df_group['Total Compensation'].groupby(level=0, group_keys=False)
```

```python
res = g.apply(lambda x: x.order(ascending=False).head())
```
