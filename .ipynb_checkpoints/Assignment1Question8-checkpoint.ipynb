```python
import nltk #importing nltk
from nltk.corpus import stopwords #importing stopwords.words('english')
from nltk.corpus import nps_chat  #importing nps_chat from nltk corpus
```

```python
file = nps_chat.fileids() # listdown all files in nps_chat
file
```

```python
chat = nps_chat.words()  #calling list of words from nps wchat
chat[0:5]
```

```python
chat = [x.lower() for x in chat] #turn all contents in lower case
```

```python
filtered_chat_words = [word for word in chat if (word.lower()) not in stopwords.words('english')] # filtering stopwords out
filtered_chat_words #len(Shake_filtered_stopwords)
```

```python
import re  #import regular expression
filter_punctuation = re.compile('.*[A-Za-z].*')  # filter out punctutations and only contain letters
filtered_webtext_punctuation = [w for w in filtered_chat_words if filter_punctuation.match(w)]
webchat_wordlist=set(filtered_webtext_punctuation)
webchat_wordlist
```

```python
from nltk.corpus import words #importing webtext from nltk corpus
```

```python
word_file = words.fileids()
word_file
```

```python
word_list = []   # create empty list
for w in word_file:           
        word1= words.words(w)
        word_list.extend(word1)
word_list = [x.lower() for x in word_list]   #lowercase all words
word_list
```

```python
unusual=[x for x in webchat_wordlist if x not in word_list] #filter out webchat words from wordlist
```

```python
# Find out how many unusual words people speak in chat 
print("Freq of unusual words: ",len(unusual))
print("Unusual words used in web chat are:")
for unusual_1 in sorted(unusual): 
    print(unusual_1)
```
